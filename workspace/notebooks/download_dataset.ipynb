```python
import os
from tqdm import tqdm
from duckduckgo_search import DDGS
from PIL import Image, UnidentifiedImageError
import requests
from io import BytesIO

# === Paramètres ===
output_root = "dataset"
num_images_per_model = 50  # nombre garanti d'images valides par modèle

# Liste de modèles (151)
models = [
    "Audi A1", "Audi A3", "Audi A4", "Audi A6", "Audi A8", "Audi Q2", "Audi Q3",
    "Audi Q5", "Audi Q7", "Audi Q8", "Audi TT", "Audi R8", "Audi e-tron",
    "BMW 1 Series", "BMW 2 Series", "BMW 3 Series", "BMW 4 Series", "BMW 5 Series",
    "BMW 7 Series", "BMW X1", "BMW X2", "BMW X3", "BMW X4", "BMW X5", "BMW X6", "BMW i3", "BMW i8",
    "Mercedes A Class", "Mercedes B Class", "Mercedes C Class", "Mercedes E Class",
    "Mercedes S Class", "Mercedes CLA", "Mercedes CLS", "Mercedes GLA", "Mercedes GLB",
    "Mercedes GLC", "Mercedes GLE", "Mercedes GLS", "Mercedes G Class", "Mercedes EQC",
    "Volkswagen Polo", "Volkswagen Golf", "Volkswagen Passat", "Volkswagen Jetta",
    "Volkswagen Arteon", "Volkswagen Tiguan", "Volkswagen Touareg", "Volkswagen T-Roc",
    "Volkswagen ID.3", "Volkswagen ID.4",
    "Toyota Yaris", "Toyota Corolla", "Toyota Camry", "Toyota Prius", "Toyota Supra",
    "Toyota RAV4", "Toyota Highlander", "Toyota Land Cruiser", "Toyota Hilux", "Toyota C-HR",
    "Honda Jazz", "Honda Civic", "Honda Accord", "Honda Insight", "Honda HR-V", "Honda CR-V",
    "Honda Pilot", "Honda Odyssey", "Honda Ridgeline", "Honda NSX",
    "Nissan Micra", "Nissan Juke", "Nissan Qashqai", "Nissan X-Trail", "Nissan Murano",
    "Nissan Pathfinder", "Nissan Navara", "Nissan GT-R", "Nissan Leaf", "Nissan 370Z",
    "Ford Fiesta", "Ford Focus", "Ford Mondeo", "Ford Mustang", "Ford Puma", "Ford Kuga",
    "Ford Edge", "Ford Explorer", "Ford Bronco", "Ford F-150",
    "Peugeot 208", "Peugeot 308", "Peugeot 508", "Peugeot 2008", "Peugeot 3008",
    "Peugeot 5008", "Peugeot Rifter", "Peugeot Traveller",
    "Renault Clio", "Renault Megane", "Renault Talisman", "Renault Zoe", "Renault Captur",
    "Renault Kadjar", "Renault Koleos", "Renault Arkana", "Renault Scenic",
    "Fiat 500", "Fiat Panda", "Fiat Tipo", "Fiat 500X", "Fiat Doblo",
    "Alfa Romeo Giulietta", "Alfa Romeo Giulia", "Alfa Romeo Stelvio", "Alfa Romeo 4C",
    "Porsche 911", "Porsche 718 Cayman", "Porsche Panamera", "Porsche Macan",
    "Porsche Cayenne", "Porsche Taycan",
    "Lamborghini Aventador", "Lamborghini Huracan", "Lamborghini Urus",
    "Ferrari 488", "Ferrari F8 Tributo", "Ferrari SF90", "Ferrari Portofino",
    "Ferrari Roma",
    "Maserati Ghibli", "Maserati Quattroporte", "Maserati Levante", "Maserati MC20",
    "Jaguar XE", "Jaguar XF", "Jaguar XJ", "Jaguar F-Pace", "Jaguar E-Pace",
    "Jaguar I-Pace", "Jaguar F-Type",
    "Land Rover Discovery", "Land Rover Defender", "Range Rover Evoque",
    "Range Rover Sport", "Range Rover Velar", "Range Rover",
    "Tesla Model S", "Tesla Model 3", "Tesla Model X", "Tesla Model Y", "Tesla Cybertruck",
    "Volvo XC40", "Volvo XC60", "Volvo XC90", "Volvo V60", "Volvo V90", "Volvo S60", "Volvo S90"
]

# Créer le dossier racine
os.makedirs(output_root, exist_ok=True)

ddgs = DDGS()

# Téléchargement avec validation stricte
for model in tqdm(models, desc="Téléchargement modèles"):
    save_dir = os.path.join(output_root, model.replace(" ", "_"))
    os.makedirs(save_dir, exist_ok=True)

    valid_count = 0
    try:
        # On demande beaucoup plus de résultats pour compenser les invalides
        results = ddgs.images(model, max_results=num_images_per_model * 10)
        for img in tqdm(results, desc=f"{model}", leave=False):
            if valid_count >= num_images_per_model:
                break  # stop quand on a assez d’images valides

            url = img.get("image")
            if not url:
                continue

            try:
                response = requests.get(url, timeout=10, stream=True)
                response.raise_for_status()

                # Vérifier que c'est bien une image
                image = Image.open(BytesIO(response.content)).convert("RGB")

                # Sauvegarder en JPG
                filename = os.path.join(save_dir, f"{model.replace(' ', '_')}_{valid_count}.jpg")
                image.save(filename, "JPEG")
                valid_count += 1

            except (requests.RequestException, UnidentifiedImageError, OSError):
                continue  # ignorer les mauvaises images

        # Vérification finale
        if valid_count < num_images_per_model:
            print(f"⚠️ Pas assez d'images pour {model} ({valid_count}/{num_images_per_model})")

    except Exception as e:
        print(f"⚠️ Erreur avec {model}: {e}")

print("✅ Téléchargement terminé avec dataset équilibré !")
```
