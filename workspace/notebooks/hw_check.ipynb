```python
import platform
import psutil
import torch

# --- Informations sur le système ---
print("=== Système ===")
print(f"Système : {platform.system()} {platform.release()}")
print(f"Architecture : {platform.machine()}")
print(f"Processeur : {platform.processor()}")

# --- Informations sur le CPU ---
print("\n=== CPU ===")
print(f"Cœurs logiques : {psutil.cpu_count(logical=True)}")
print(f"Cœurs physiques : {psutil.cpu_count(logical=False)}")
print(f"Fréquence max : {psutil.cpu_freq().max:.2f} MHz")
print(f"Utilisation CPU globale : {psutil.cpu_percent(interval=1)} %")

# --- Informations sur la RAM ---
mem = psutil.virtual_memory()
print("\n=== RAM ===")
print(f"Total : {mem.total / (1024 ** 3):.2f} GB")
print(f"Disponible : {mem.available / (1024 ** 3):.2f} GB")
print(f"Utilisée : {mem.used / (1024 ** 3):.2f} GB")
print(f"Pourcentage utilisé : {mem.percent} %")

# --- Informations sur le GPU NVIDIA via PyTorch ---
print("\n=== GPU NVIDIA (CUDA) ===")
if torch.cuda.is_available():
    num_gpus = torch.cuda.device_count()
    print(f"Nombre de GPU détectés : {num_gpus}")
    for i in range(num_gpus):
        print(f"GPU {i} : {torch.cuda.get_device_name(i)}")
        print(f"  Mémoire totale : {torch.cuda.get_device_properties(i).total_memory / (1024**3):.2f} GB")
else:
    print("Aucun GPU NVIDIA détecté ou CUDA non disponible")
```
