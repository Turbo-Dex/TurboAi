```python
import os

labels_folders = ["stanford_dataset/train/labels",
                  "stanford_dataset/valid/labels",
                  "stanford_dataset/val/labels"]

class_ids = set()

for folder in labels_folders:
    for file in os.listdir(folder):
        if file.endswith(".txt"):
            with open(os.path.join(folder, file), "r") as f:
                for line in f:
                    if line.strip():
                        class_id = int(line.split()[0])
                        class_ids.add(class_id)

# Trier les classes
class_ids = sorted(list(class_ids))
print(f"Nombre de classes : {len(class_ids)}")
print(f"IDs de classes : {class_ids}")
```
